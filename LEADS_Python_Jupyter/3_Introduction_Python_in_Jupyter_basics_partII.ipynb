***You can hold multiple values using: lists, tuples, dictionaries***
***

* **Lists** - a list of values. Each one of them is numbered, starting from zero - the first one is numbered zero, the second 1, the third 2, etc. You can remove values from the list, and add new values to the end. Example: your study sites.
* **Tuples** - are just like lists, but you can't change their values. The initial values are fixed for the rest of the program. Again, each value is numbered starting from zero, for easy reference. Example: the names of the months of the year.
* **Dictionaries** - in a dictionary, you have an 'index' of words, and for each of them a definition. In Python, the word is called a 'key', and the definition a 'value'. The values in a dictionary aren't numbered - they aren't in any specific order, either - the key does the same thing. You can add, remove, and modify the values in dictionaries. Example: telephone book.

## 3.2 Sequence of instructions, loops, branches (or selection)
***

Programming code can often be broken down into three main programming structures:
- sequence
- branches (or selection)
- loops

***
### *3.2.1 A sequence is a series of actions completed in a specific order (following a specific path), like making a cup of tea* <br>
***

**Or: opening the data > storing it in a variable > manipulating the data > printing the output**



````python
# You always have to plan your code
# Example of a sequence of instructions

x = int(input('What is the price of a cupcake?',)) # input and storage of data
y = int(input('How many cupcakes do we need?',)) # input and storage of data
Cost = x*y # manupilation of data
print('The total cost of the cupcakes is',Cost) # output of manupilated data
````

***
### *3.2.2 Branches or selections offer different paths depending on conditions that are met. Like deciding to bike to school when it doesn't rains, but drive when it does rain.* <br>
***

**Branches ask a question, for example: is the coordinate system of ImageA equal to WGS 84? If true > read the next file. If false > project the file to WGS 84**



**IF-ELSE-THEN control structures are used to implement branches**

The `if` statement, along with `else` and `elif` follow this form:
```Python
if {conditions}:
    {run this code}
elif {conditions}:
    {run this code}
elif {conditions}:
    {run this code}
else:
    {run this code}
    
#You can have as many or as little elif statements as you need
#anywhere from zero to the sky.
#You can have at most one else statement
#and only after all other ifs and elifs.
```

````python
# Example of simple IF structure, only one condition is verified
# Define a function to calculate the factors of a number
# The name of the function is get_factors

def get_factors1(num):
    df=open('factorsCorrect.txt','w')
    for i in range(1, num + 1):
        if num % i == 0:
            print("{} is a factor of {}.".format(i, num))
            df.write(str(i))
            df.write('\n')
    df.close    
````

````python
get_factors1(12)
````

***
### *3.2.3 Loops repeat the same action (s) over and over again, until a condition is met. Like checking the water on the stove, over and over again, until it boils. Then you take the pot of the fire to make your cup of tea.* <br>
***

**Loops ask the same question, for example: are all the images clipped to their study area? No, then read the next image and clip it. Until all images are clipped.**


**In python you can use the `while` and `for` control structures to implement loops**

````python
#While loop example
a = 0
while a < 10:
    a = a + 1
    print(a)
````

````python
# Example of simple IF structure, only one condition is verified
# Also example of FOR loop, loops through the number you choose, at steps of 1
# Define a function to calculate the factors of a number
# The name of the function is get_factors

def get_factors1(num):
    df=open('factorsCorrect.txt','w')
    for i in range(1, num + 1):
        if num % i == 0:
            print("{} is a factor of {}.".format(i, num))
            df.write(str(i))
            df.write('\n')
    df.close    

get_factors1(8)
````
