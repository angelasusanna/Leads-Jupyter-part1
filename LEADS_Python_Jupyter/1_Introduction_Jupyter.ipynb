**A Jupyter notebook consists of cells that can store text (and figures, equations,etc) or code. There are 3 types of cells:**
1. Code cells - this is where you will write your python commands
2. Markdown cells- these cells can hold explanatory text and figures
3. Raw cells - these are cells with no formatting and are of use if your intend to convert your notebook to another format using the command line too nbconvert
